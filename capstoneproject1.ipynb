```python
import pandas as pd

# URL del archivo CSV para los presupuestos de bomberos
url_fire_budget = "https://raw.githubusercontent.com/erikaguiracocha/Data-Visualization-2025/refs/heads/main/Fire%20budgets.csv"

# Cargar el archivo CSV en un DataFrame
df_fire_budget = pd.read_csv(url_fire_budget)

# Limpiar los valores eliminando las comas y convertir a tipo numérico
for column in df_fire_budget.columns[1:]:  # Omitir la columna 'Year'
    df_fire_budget[column] = df_fire_budget[column].replace({',': ''}, regex=True)  # Eliminar las comas
    df_fire_budget[column] = pd.to_numeric(df_fire_budget[column], errors='coerce')  # Convertir a numérico

# Verificar las primeras filas de los datos limpiados
print(df_fire_budget.head())

# Establecer el formato de moneda para las salidas
def format_currency(x):
    return f"${x:,.2f}"

# Calcular las estadísticas descriptivas
stats = {}

# Obtener el nombre de las ciudades
cities = df_fire_budget.columns[1:]

# Calcular la media, mediana, moda, máximo y mínimo por cada ciudad
for city in cities:
    stats[city] = {
        'Average': format_currency(df_fire_budget[city].mean()),
        'Median': format_currency(df_fire_budget[city].median()),
        'Mode': format_currency(df_fire_budget[city].mode()[0]),  # Moda (consideramos el primer valor si hay más de una)
        'Max': format_currency(df_fire_budget[city].max()),
        'Min': format_currency(df_fire_budget[city].min())
    }

# Convertir el resultado a un DataFrame para presentación más clara
df_stats = pd.DataFrame(stats)

# Mostrar las estadísticas descriptivas
df_stats
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# URLs para los datasets
fire_budget_url = "https://raw.githubusercontent.com/erikaguiracocha/Data-Visualization-2025/refs/heads/main/Fire%20budgets.csv"
police_budget_url = "https://raw.githubusercontent.com/erikaguiracocha/Data-Visualization-2025/refs/heads/main/Police%20budgets.csv"
response_time_url = "https://raw.githubusercontent.com/erikaguiracocha/Data-Visualization-2025/refs/heads/main/time%20of%20responses.csv"

# Cargar los datasets
fire_budget_data = pd.read_csv(fire_budget_url)
police_budget_data = pd.read_csv(police_budget_url)
response_time_data = pd.read_csv(response_time_url)

# Limpiar los nombres de las columnas (eliminar espacios en blanco)
fire_budget_data.columns = fire_budget_data.columns.str.strip()
police_budget_data.columns = police_budget_data.columns.str.strip()
response_time_data.columns = response_time_data.columns.str.strip()

# Corregir errores tipográficos en los nombres de las ciudades
response_time_data = response_time_data.rename(columns={'Hamilton Towship, NJ': 'Hamilton Township, NJ'})

# Convertir 'Year' a tipo numérico y eliminar filas con valores no numéricos
def clean_year_column(df):
    df = df[pd.to_numeric(df['Year'], errors='coerce').notnull()]  # Eliminar filas con valores no numéricos
    df['Year'] = df['Year'].astype(int)  # Convertir la columna 'Year' a tipo entero
    return df

# Limpiar todos los datasets
fire_budget_data = clean_year_column(fire_budget_data)
police_budget_data = clean_year_column(police_budget_data)
response_time_data = clean_year_column(response_time_data)

# Función para limpiar las columnas de presupuesto y convertirlas a float
def clean_budget_columns(df, budget_columns):
    for col in budget_columns:
        df[col] = df[col].replace({',': ''}, regex=True)  # Eliminar comas
        df[col] = df[col].astype(float)  # Convertir a float
    return df

# Limpiar columnas de presupuesto
fire_budget_columns = ['Hamilton Township, NJ', 'Trenton, NJ', 'Ewing, NJ', 'Morris Township, NJ', 'Newark, NJ']
police_budget_columns = ['Hamilton Township, NJ', 'Trenton, NJ', 'Ewing, NJ', 'Morris Township, NJ', 'Newark, NJ']

# Limpiar presupuestos
fire_budget_data = clean_budget_columns(fire_budget_data, fire_budget_columns)
police_budget_data = clean_budget_columns(police_budget_data, police_budget_columns)

# Función para convertir el tiempo en formato 'hh:mm' a minutos
def convert_time_to_minutes(df, time_columns):
    for col in time_columns:
        df[col] = df[col].apply(lambda x: int(x.split(':')[0])*60 + int(x.split(':')[1]))  # Convertir a minutos
    return df

# Columnas de tiempo de respuesta
time_columns = ['Hamilton Township, NJ', 'Trenton, NJ', 'Ewing, NJ', 'Morris Township, NJ', 'Newark, NJ']

# Convertir los tiempos de respuesta a minutos
response_time_data = convert_time_to_minutes(response_time_data, time_columns)

# Fusionar los datasets en la columna 'Year'
merged_data = pd.merge(fire_budget_data, police_budget_data, on='Year', suffixes=('_fire', '_police'))
merged_data = pd.merge(merged_data, response_time_data, on='Year')

# Calcular la correlación entre las variables
correlation = merged_data[['Hamilton Township, NJ_fire', 'Trenton, NJ_fire', 'Ewing, NJ_fire', 'Morris Township, NJ_fire', 'Newark, NJ_fire',
                           'Hamilton Township, NJ_police', 'Trenton, NJ_police', 'Ewing, NJ_police', 'Morris Township, NJ_police', 'Newark, NJ_police',
                           'Hamilton Township, NJ', 'Trenton, NJ', 'Ewing, NJ', 'Morris Township, NJ', 'Newark, NJ']].corr()

# Crear el mapa de calor
plt.figure(figsize=(12, 8))
sns.heatmap(correlation, annot=True, cmap='coolwarm', center=0, fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap between Fire and Police Budget and Response Times by City')
plt.show()
```

#**Visualization calls and response times**

```python
import pandas as pd
import matplotlib.pyplot as plt

# URLs for volume of calls and response times
volume_url = "https://raw.githubusercontent.com/erikaguiracocha/Data-Visualization-2025/refs/heads/main/volume%20of%20calls.csv"
response_time_url = "https://raw.githubusercontent.com/erikaguiracocha/Data-Visualization-2025/refs/heads/main/time%20of%20responses.csv"

# Read the CSV files from the URLs
data_calls = pd.read_csv(volume_url)
data_response = pd.read_csv(response_time_url)

# Rename columns to match city names for consistency
data_calls = data_calls.rename(columns={'Hamilton Towship, NJ': 'Hamilton Township, NJ'})

# Reshape the data for easier comparison
# Volume of Calls
calls_melted = data_calls.melt(id_vars=['Year'], var_name='City', value_name='Volume of Calls')

# Time of Response: Convert time (HH:MM) into minutes for comparison
def convert_time_to_minutes(time_str):
    minutes, seconds = time_str.split(':')
    return int(minutes) + int(seconds) / 60

data_response_melted = data_response.rename(columns={'Hamilton Towship, NJ': 'Hamilton Township, NJ'})
data_response_melted = data_response_melted.melt(id_vars=['Year'], var_name='City', value_name='Response Time')
data_response_melted['Response Time'] = data_response_melted['Response Time'].apply(convert_time_to_minutes)

# Merge both datasets on 'City'
merged_data = pd.merge(calls_melted, data_response_melted, on=['Year', 'City'])

# Calculate the average volume of calls and response time per city
city_comparison = merged_data.groupby('City').agg({
    'Volume of Calls': 'mean',
    'Response Time': 'mean'
}).reset_index()

# Plotting the comparison for Volume of Calls and Response Time
fig, ax1 = plt.subplots(figsize=(12, 8))

# Set bar width
bar_width = 0.35

# Set positions for the bars
index = range(len(city_comparison))

# Create the bars for Volume of Calls on the left Y-axis
ax1.bar(index, city_comparison['Volume of Calls'], bar_width, label='Volume of Calls', color='blue')

# Create a second Y-axis for Response Time
ax2 = ax1.twinx()

# Create the bars for Response Time on the right Y-axis
ax2.bar([i + bar_width for i in index], city_comparison['Response Time'], bar_width, label='Response Time (minutes)', color='red')

# Customize the plot
ax1.set_xlabel('City', fontsize=12)
ax1.set_ylabel('Volume of Calls', fontsize=12, color='blue')
ax2.set_ylabel('Response Time (minutes)', fontsize=12, color='red')
ax1.set_title('Comparison of Volume of Calls and Response Time by City', fontsize=14)

# Set the cities on the X-axis
ax1.set_xticks([i + bar_width / 2 for i in index])
ax1.set_xticklabels(city_comparison['City'], rotation=45)

# Set color for the ticks on both axes
ax1.tick_params(axis='y', labelcolor='blue')
ax2.tick_params(axis='y', labelcolor='red')

# Add Legends
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Show plot
plt.tight_layout()
plt.show()
```
