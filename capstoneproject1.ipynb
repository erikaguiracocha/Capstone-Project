#Uploaded police budgets

```python
import pandas as pd
from tabulate import tabulate

# Cargar el archivo CSV desde el URL de GitHub
url = "https://raw.githubusercontent.com/erikaguiracocha/Data-Visualization-2025/refs/heads/main/Police%20budgets.csv"
df_police_budget = pd.read_csv(url)

# Asegurarse de que la columna 'Year' sea de tipo numérico para realizar el filtro
df_police_budget['Year'] = pd.to_numeric(df_police_budget['Year'], errors='coerce')

# Filtrar los datos para los años de 2013 a 2023
df_filtered = df_police_budget[(df_police_budget['Year'] >= 2013) & (df_police_budget['Year'] <= 2023)]

# Mostrar el DataFrame filtrado de forma bonita usando tabulate
print(tabulate(df_filtered, headers='keys', tablefmt='pretty', showindex=False))
```

```python
import pandas as pd

# URL del archivo CSV en GitHub
url = "https://raw.githubusercontent.com/erikaguiracocha/Data-Visualization-2025/refs/heads/main/Police%20budgets.csv"

# Cargar los datos desde el archivo CSV en GitHub
df_police_budget = pd.read_csv(url)

# Eliminar las comas de los valores para convertirlos a números
df_police_budget = df_police_budget.replace({',': ''}, regex=True)

# Convertir las columnas a tipo numérico
df_police_budget = df_police_budget.apply(pd.to_numeric, errors='coerce')

# Ver las primeras filas para revisar los datos
df_police_budget.head()

# Estadísticas descriptivas por ciudad
cities = ['Hamilton Township, NJ', 'Trenton, NJ', 'Ewing, NJ', 'Morris Township, NJ', 'Newark, NJ']

# Calculando estadísticas
stats = {}
for city in cities:
    stats[city] = {
        'Average': df_police_budget[city].mean(),
        'Median': df_police_budget[city].median(),
        'Mode': df_police_budget[city].mode()[0],  # Moda (consideramos el primer valor si hay más de una)
        'Max': df_police_budget[city].max(),
        'Min': df_police_budget[city].min()
    }

# Convertir el diccionario de estadísticas a un DataFrame para mejor presentación
stats_df = pd.DataFrame(stats)

# Ajustar el formato a dólares
pd.options.display.float_format = '${:,.2f}'.format

# Mostrar las estadísticas
stats_df
```

```python
import pandas as pd

# URL del archivo CSV para los presupuestos de bomberos
url_fire_budget = "https://raw.githubusercontent.com/erikaguiracocha/Data-Visualization-2025/refs/heads/main/Fire%20budgets.csv"

# Cargar el archivo CSV en un DataFrame
df_fire_budget = pd.read_csv(url_fire_budget)

# Limpiar los valores eliminando las comas y convertir a tipo numérico
for column in df_fire_budget.columns[1:]:  # Omitir la columna 'Year'
    df_fire_budget[column] = df_fire_budget[column].replace({',': ''}, regex=True)  # Eliminar las comas
    df_fire_budget[column] = pd.to_numeric(df_fire_budget[column], errors='coerce')  # Convertir a numérico

# Verificar las primeras filas de los datos limpiados
print(df_fire_budget.head())

# Establecer el formato de moneda para las salidas
def format_currency(x):
    return f"${x:,.2f}"

# Calcular las estadísticas descriptivas
stats = {}

# Obtener el nombre de las ciudades
cities = df_fire_budget.columns[1:]

# Calcular la media, mediana, moda, máximo y mínimo por cada ciudad
for city in cities:
    stats[city] = {
        'Average': format_currency(df_fire_budget[city].mean()),
        'Median': format_currency(df_fire_budget[city].median()),
        'Mode': format_currency(df_fire_budget[city].mode()[0]),  # Moda (consideramos el primer valor si hay más de una)
        'Max': format_currency(df_fire_budget[city].max()),
        'Min': format_currency(df_fire_budget[city].min())
    }

# Convertir el resultado a un DataFrame para presentación más clara
df_stats = pd.DataFrame(stats)

# Mostrar las estadísticas descriptivas
df_stats
```
