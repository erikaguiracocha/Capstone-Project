```python
import pandas as pd

# URL del archivo CSV para los presupuestos de bomberos
url_fire_budget = "https://raw.githubusercontent.com/erikaguiracocha/Data-Visualization-2025/refs/heads/main/Fire%20budgets.csv"

# Cargar el archivo CSV en un DataFrame
df_fire_budget = pd.read_csv(url_fire_budget)

# Limpiar los valores eliminando las comas y convertir a tipo numérico
for column in df_fire_budget.columns[1:]:  # Omitir la columna 'Year'
    df_fire_budget[column] = df_fire_budget[column].replace({',': ''}, regex=True)  # Eliminar las comas
    df_fire_budget[column] = pd.to_numeric(df_fire_budget[column], errors='coerce')  # Convertir a numérico

# Verificar las primeras filas de los datos limpiados
print(df_fire_budget.head())

# Establecer el formato de moneda para las salidas
def format_currency(x):
    return f"${x:,.2f}"

# Calcular las estadísticas descriptivas
stats = {}

# Obtener el nombre de las ciudades
cities = df_fire_budget.columns[1:]

# Calcular la media, mediana, moda, máximo y mínimo por cada ciudad
for city in cities:
    stats[city] = {
        'Average': format_currency(df_fire_budget[city].mean()),
        'Median': format_currency(df_fire_budget[city].median()),
        'Mode': format_currency(df_fire_budget[city].mode()[0]),  # Moda (consideramos el primer valor si hay más de una)
        'Max': format_currency(df_fire_budget[city].max()),
        'Min': format_currency(df_fire_budget[city].min())
    }

# Convertir el resultado a un DataFrame para presentación más clara
df_stats = pd.DataFrame(stats)

# Mostrar las estadísticas descriptivas
df_stats
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load datasets
url_police = "https://raw.githubusercontent.com/erikaguiracocha/Data-Visualization-2025/refs/heads/main/Police%20budgets.csv"
url_fire = "https://raw.githubusercontent.com/erikaguiracocha/Data-Visualization-2025/refs/heads/main/Fire%20budgets.csv"
url_calls = "https://raw.githubusercontent.com/erikaguiracocha/Data-Visualization-2025/refs/heads/main/call%20volumes%20in%20cvs.csv"

df_police = pd.read_csv(url_police)
df_fire = pd.read_csv(url_fire)
df_calls = pd.read_csv(url_calls)

# Clean column names
df_police.columns = df_police.columns.str.strip()
df_fire.columns = df_fire.columns.str.strip()
df_calls.columns = df_calls.columns.str.strip()

# Convert 'Year' to string in all datasets
df_police['Year'] = df_police['Year'].astype(str)
df_fire['Year'] = df_fire['Year'].astype(str)
df_calls['Year'] = df_calls['Year'].astype(str)

# Convert police and fire data to long format
df_police_long = df_police.melt(id_vars='Year', var_name='City', value_name='Police_Budget')
df_fire_long = df_fire.melt(id_vars='Year', var_name='City', value_name='Fire_Budget')

# Clean numeric columns
df_police_long['Police_Budget'] = df_police_long['Police_Budget'].str.replace(',', '').astype(float)
df_fire_long['Fire_Budget'] = df_fire_long['Fire_Budget'].str.replace(',', '').astype(float)
df_calls['Estimated Total Calls'] = df_calls['Estimated Total Calls'].str.replace(',', '').astype(float)

# Rename column
df_calls.rename(columns={'Estimated Total Calls': 'Call_Volume'}, inplace=True)

# Standardize city names
df_police_long['City'] = df_police_long['City'].str.replace(', NJ', '', regex=False).str.strip()
df_fire_long['City'] = df_fire_long['City'].str.replace(', NJ', '', regex=False).str.strip()
df_calls['City'] = df_calls['City'].str.strip()

# Merge all three datasets
df_merged = pd.merge(df_police_long, df_fire_long, on=['Year', 'City'])
df_merged = pd.merge(df_merged, df_calls, on=['Year', 'City'])

# Calculate correlation matrix
correlation = df_merged[['Police_Budget', 'Fire_Budget', 'Call_Volume']].corr()

# Display correlation matrix as table
print("Correlation Table:")
print(correlation)

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation between Budgets and Call Volume")
plt.show()
```

#Average of call volumes graph

```python
import pandas as pd
import matplotlib.pyplot as plt

# URL of the CSV file
url_calls = "https://raw.githubusercontent.com/erikaguiracocha/Data-Visualization-2025/refs/heads/main/call%20volumes%20in%20cvs.csv"

# Load the data
df_calls = pd.read_csv(url_calls)

# Clean column names
df_calls.columns = df_calls.columns.str.strip()

# Rename for clarity
df_calls = df_calls.rename(columns={"Estimated Total Calls": "Call Volume"})

# Convert to numeric (remove commas if present)
df_calls["Call Volume"] = pd.to_numeric(df_calls["Call Volume"].astype(str).str.replace(",", ""), errors='coerce')

# Average by city
city_avg_calls = df_calls.groupby("City")["Call Volume"].mean().reset_index()
city_avg_calls = city_avg_calls.sort_values("City").reset_index(drop=True)

# Plot
plt.figure(figsize=(10, 6))
bars = plt.bar(city_avg_calls["City"], city_avg_calls["Call Volume"], color="skyblue")

# Add labels
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 20, f'{height:.0f}', ha='center', fontsize=9)

# Labels and title
plt.xlabel("City")
plt.ylabel("Average Call Volume")
plt.title("Average Call Volume by City (2013–2023)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# URL del archivo CSV de response time
url_response = "https://raw.githubusercontent.com/erikaguiracocha/Data-Visualization-2025/refs/heads/main/response%20times%20en%20cvs.csv"

# Cargar los datos
df_response = pd.read_csv(url_response)

# Limpiar nombres de columnas
df_response.columns = df_response.columns.str.strip()

# Renombrar columna por claridad
df_response = df_response.rename(columns={"90th Percentile Response Time": "Response Time"})

# Convertir tiempos MM:SS a minutos
def convert_to_minutes(t):
    try:
        minutes, seconds = t.strip().split(":")
        return int(minutes) + int(seconds) / 60
    except:
        return None

df_response["Response Time"] = df_response["Response Time"].apply(convert_to_minutes)

# Calcular promedio por ciudad
city_avg_response = df_response.groupby("City")["Response Time"].mean().reset_index()
city_avg_response = city_avg_response.sort_values("City").reset_index(drop=True)

# Graficar
plt.figure(figsize=(10, 6))
bars = plt.bar(city_avg_response["City"], city_avg_response["Response Time"], color="salmon")

# Agregar etiquetas
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.1, f'{height:.2f}', ha='center', fontsize=9)

# Etiquetas y título
plt.xlabel("City")
plt.ylabel("Average Response Time (minutes)")
plt.title("Average Emergency Response Time by City (2013–2023)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Supongamos que tienes un DataFrame llamado df con las siguientes columnas:
# 'Fire Budget', 'Police Budget', 'Population', 'Income', 'Unemployment'

# Configuración de estilo para los gráficos
sns.set(style="whitegrid")

# Crear una figura con 3 subgráficos (uno para cada variable socioeconómica)
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Gráfico de dispersión: Fire Budget vs Population
sns.scatterplot(x='Population', y='Fire Budget', data=df, ax=axes[0], color='blue')
axes[0].set_title('Fire Budget vs Population')
axes[0].set_xlabel('Population')
axes[0].set_ylabel('Fire Budget')

# Gráfico de dispersión: Fire Budget vs Income
sns.scatterplot(x='Income', y='Fire Budget', data=df, ax=axes[1], color='green')
axes[1].set_title('Fire Budget vs Income')
axes[1].set_xlabel('Income')
axes[1].set_ylabel('Fire Budget')

# Gráfico de dispersión: Fire Budget vs Unemployment
sns.scatterplot(x='Unemployment', y='Fire Budget', data=df, ax=axes[2], color='red')
axes[2].set_title('Fire Budget vs Unemployment')
axes[2].set_xlabel('Unemployment')
axes[2].set_ylabel('Fire Budget')

# Ajustar el espaciado entre los subgráficos
plt.tight_layout()

# Mostrar los gráficos
plt.show()
```

```python
# Crear una figura con 3 subgráficos (uno para cada variable socioeconómica)
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Gráfico de dispersión: Police Budget vs Population
sns.scatterplot(x='Population', y='Police Budget', data=df, ax=axes[0], color='blue')
axes[0].set_title('Police Budget vs Population')
axes[0].set_xlabel('Population')
axes[0].set_ylabel('Police Budget')

# Gráfico de dispersión: Police Budget vs Income
sns.scatterplot(x='Income', y='Police Budget', data=df, ax=axes[1], color='green')
axes[1].set_title('Police Budget vs Income')
axes[1].set_xlabel('Income')
axes[1].set_ylabel('Police Budget')

# Gráfico de dispersión: Police Budget vs Unemployment
sns.scatterplot(x='Unemployment', y='Police Budget', data=df, ax=axes[2], color='red')
axes[2].set_title('Police Budget vs Unemployment')
axes[2].set_xlabel('Unemployment')
axes[2].set_ylabel('Police Budget')

# Ajustar el espaciado entre los subgráficos
plt.tight_layout()

# Mostrar los gráficos
plt.show()
```

#**Graphs of times of responses and budgets**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker

# Example data for budgets and response times by city
data = {
    'City': ['Hamilton Township, NJ', 'Trenton, NJ', 'Ewing, NJ', 'Morris Township, NJ', 'Newark, NJ'],
    'Fire Budget': [1000000, 980000, 940000, 970000, 990000],  # Fire Budget in dollars
    'Police Budget': [1000000, 970000, 950000, 960000, 990000],  # Police Budget in dollars
    'Response Time (Fire)': [5.2, 6.1, 5.8, 5.3, 5.5],  # Response times in minutes
    'Response Time (Police)': [4.8, 5.2, 5.0, 4.9, 5.3]  # Response times in minutes
}

# Create the DataFrame
df = pd.DataFrame(data)

# Set the style for the graphs
sns.set(style="whitegrid")

# Create a bar plot for the budgets
plt.figure(figsize=(10, 6))
ax = df.set_index('City')[['Fire Budget', 'Police Budget']].plot(kind='bar', width=0.8, color=['blue', 'red'])

# Title and labels
plt.title('Comparison of Fire and Police Budgets by City', fontsize=16)
plt.xlabel('City', fontsize=12)
plt.ylabel('Budget ($)', fontsize=12)

# Set Y-axis to show values in dollar format with ranges
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f'${x:,.0f}'))

# Adjust X-axis labels to prevent overlapping
plt.xticks(rotation=45, ha='right', fontsize=10)

# Add a legend
plt.legend(title='Budget', bbox_to_anchor=(1.05, 1), loc='upper left')

# Adjust layout to avoid overlap
plt.tight_layout()

# Show the plot
plt.show()
```
